```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.read_csv('../data/TUANDROMD.csv')
vals = data.values[:, :data.shape[1] - 1]
```

```python
clf = LocalOutlierFactor(n_neighbors=10, contamination=0.1)
y_pred = clf.fit_predict(vals)
a = sum([1 for a in y_pred if a == -1])
print('Detected ' + str(a) + ' outliers using Local Outlier Factor')
clean_data_LOF = pd.DataFrame(columns=data.columns)
for val_idx in range(len(vals)):
    if y_pred[val_idx] == 1:
        clean_data_LOF.loc[val_idx] = data.loc[val_idx]
clean_data_LOF.head(5)
```

```python
clean_data_LOF.to_csv('../data/cleanLOF.csv', index=False)
```
